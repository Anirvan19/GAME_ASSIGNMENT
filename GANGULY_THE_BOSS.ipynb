```python
import sys
import pygame
from pygame.locals import *
pygame.init()
pygame.mixer.init()
FPS = 120
fpsClock = pygame.time.Clock()
screen_width = 450
screen_height = 450
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption("GANGULY_THE_BOSS")
cyan = pygame.Color(0, 255, 255) #setting colors up for later use in variables.
purple = pygame.Color(128, 0, 128)
BUTTON_bg_COLOR = pygame.Color(68, 93, 255)
BUTTON_BORDER_COLOR = pygame.Color(85, 50, 232)
FONT = pygame.font.SysFont("ariafont.txt", 50)
gangu_IMAGE = pygame.image.load("download.jpeg")
ganguly = 0
def draw(text, font, color, surface, x, y):
    text_surface = font.render(text, True, color)
    text_rect = text_surface.get_rect()
    text_rect.topleft = (x, y)
    surface.blit(text_surface, text_rect)
class Item: #OOPS part, here we play with object variables
    def __init__(self, rect, text, base_price, base_cps_each):
        self.rect = rect
        self.text = text
        self.count = 0
        self.base_price = base_price
        self.cps_each = base_cps_each
    def draw(self, surface):
        pygame.draw.rect(surface, BUTTON_bg_COLOR, self.rect, 0)
        pygame.draw.rect(surface, BUTTON_bg_COLOR, self.rect, 2)
        text_surface = FONT.render(
            f"{self.count}x{self.text} ${int(self.price())}", False, cyan
        )
        text_rect = text_surface.get_rect()
        text_rect.topleft = (
            self.rect.left + 10,
            self.rect.top + self.rect.height * 0.25,
        )
        surface.blit(text_surface, text_rect)
    def total_cps(self):
        return self.cps_each * self.count
    def price(self):
        return self.base_price * 1.15 ** self.count
    def click(self):
        price = self.price()
        global ganguly
        if ganguly >= price:
            self.count += 1
            ganguly -= price
    def collidepoint(self, point):
        return self.rect.collidepoint(point)
gangu_rect = Rect(25, 250, gangu_IMAGE.get_width(), gangu_IMAGE.get_height())
def click_gangu():
    global ganguly
    ganguly += 1
def update_ganguly():
    global ganguly
elapsed_time = 0
start_time = pygame.time.get_ticks()
time_limit = 15
clicks = 0


#{GAME LOOP}
while True:
    screen.fill(cyan)
    screen.blit(gangu_IMAGE, gangu_rect)

    #for the timer
    elapsed_time = (pygame.time.get_ticks() - start_time) / 1000
    remaining_time = max(0, time_limit - elapsed_time)
    draw(f"Time: {int(remaining_time)} seconds",FONT,purple,screen,10,10,)
    draw(str(int(ganguly)), FONT, purple, screen, 100, 200)

    if remaining_time <= 0:
        draw(f"Clicks: {clicks}",FONT,purple,screen,screen_width // 2 - 50,screen_height // 2,)
    else:
        for event in pygame.event.get():
            if event.type == MOUSEBUTTONDOWN:
                mouse_pos = event.pos
                mouse_button = event.button
                if mouse_button == 1:
                    if gangu_rect.collidepoint(mouse_pos):
                        click_gangu()
                        clicks += 1
    pygame.display.update()
    fpsClock.tick(FPS)
    if elapsed_time >= time_limit:
        break
while True:
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()

    pygame.display.update()
    fpsClock.tick(FPS)
pygame.quit()
```
